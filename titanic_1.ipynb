```python
females = data.loc[data["Sex"] == 'female']['Name'].tolist()
```

```python
females
```

```python
for i in range(len(females)):
    if '(' in females[i]:
        females[i] = females[i].split('(')[-1]
    elif 'Mrs.' in females[i]:
        females[i] = females[i].split('Mrs. ')[1]
    elif 'Miss' in females[i]:
        females[i] = females[i].split('Miss. ')[1]
    females[i] = females[i].split(' ')[0].replace(')', '')
```

```python
f = list(set(females))
b = sorted(f, key=lambda x:females.count(x))
print(b)
print(b[-1])
```
